```python
#Visualisation

sns.histplot(df.age)
plt.title('Age distribution of the patient')
plt.show()
```

```python
#Chest pain type

sns.histplot(df.cp)
plt.title('Patient who has chest pain')
plt.show()
```

```python
#Resting blood pressure

sns.histplot(df.trtbps)
plt.title('Blood Pressure Patient')
plt.show()

#We observe that blood pressure has a high peak at a value of approx. 150
```

```python
#Cholesterol

sns.histplot(df.chol)
plt.title('Cholesterol Level Distribution')
plt.show()

#In our dataset we observe that patient have extremely high level of cholesterol.
```

```python
#fasting blood sugar

df.fbs.value_counts()

# 0 Menas less or equal to 120 mg/dl (which is good) and 1 means higher than 120 mg/dl.
```

```python
#Electrocardiographic Test Distribution

sns.histplot(df.restecg, bins = 3)
sns.histplot()
plt.title('Electrocardiographic test Distribution')
plt.show()
```

```python
#Observation - 0 - normal , 1 - hypertrophy, 2 - having ST- T wave abnormality. According to the graph - about 50% have hypertrophy and few patient have ST-T wave abnormality, rets of them have normal result.
```

```python
#Maximum heart rate achieved 

sns.histplot(df.thalachh)
plt.title ('Maximum heart rate')
plt.show()
```

```python
#observation - Highest heart rate between 170-180 
```

```python
#the chances of heart Attack 
df.output.value_counts()
```

```python
#0 less chances and 1 high chances of heart Attack so 164 patient have high chances of Heart Attack and 138 patient have low chances of heart Attack.
```

```python
#Conclusion - That people with good health have less chances of heart Attack
```

```python
#Correlation - Statistical Relationship between 2 variable . Can't measure in units and useful for identifying relationship
#between  Variable.

#Finding Correlation
plt.figure(figsize = (20,9))
sns.heatmap(df.corr(), annot= True);
plt.title('correlation map')
```

```python
#observation - We observe that 'cp', 'thalalchh' and 'slp' have highly correlated to output.
```

```python
#Pairplot to Numeric columns 

sns.pairplot(df[['age', 'trtbps', 'chol', 'oldpeak', 'output']], hue = 'output')
```

```python
#Observation 
# 1. The Majority of the patient are older than 40.
# 2. In the dataset there are 2 times more males than females.
# 3. 54.3% high chances of heart attack and 45.7% have less chances of heart attack.
# 4. Age distribution of the patient from high chances group has a peak at 57-58 ages.
# 5. Patient have normal analysis result (like electrocardiography and thallium stress test) are less likely to heart attack.
# 6. The Major factor influencing heart attack are : chest pain, Maximum heart rate, Exercise induced Angina, ST Depression induced by Exercise relative to rest and Number of major vessels. 
```
